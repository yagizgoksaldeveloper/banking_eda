```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the csv file
df = pd.read_csv("bank.csv" , sep = ";")
```

```python
# General look to the DataFrame
df
```

```python
# Display a summary of the dataset, including column data types and non-null values
df.info()
```

```python
# Count the number of missing values in each column of the dataset
df.isnull().sum()
```

```python
# Generate a statistical summary of the dataset's numerical features
df.describe()
```

```python
# Count the number of term deposit subscriptions and non-subscriptions in the dataset
df["y"].value_counts()
```

```python
# Visualize the distribution of term deposit subscriptions in the dataset
df["y"].value_counts().plot(kind = "bar")
plt.xlabel("Term Deposit Subscription")
plt.ylabel("Count")
plt.show()
```

```python
# Look for the Term Deposit Subscription distribution according to age.
df.boxplot(column = "age", by = "y")
plt.xlabel("Term Deposit Subscription")
plt.ylabel("Age")
plt.show()
```

```python
# Generate a correlation matrix and visualize it as a heatmap
corr_matrix = df.corr()
plt.matshow(corr_matrix)
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation='vertical')
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.colorbar()
plt.show()
```
